```r
ages <- c(1, 3, 5, 2, 11, 9, 3, 9, 12, 3)
weights <- c(4.4, 5.3, 7.2, 5.2, 8.5, 7.3, 6.0, 10.4, 10.2, 6.1)
cat(sprintf("mean: %f\n", mean(weights)))
cat(sprintf("standard deviation: %f\n", sd(weights)))
cat(sprintf("correlation: %f\n", cor(ages, weights)))
plot(ages, weights)
```

```r
indices <- c(121, 121, 122, 123, 123, 121, 121, 122)
probabilities <- c(96.5, 43, 22.1, 1.4, 89, 99, 34, 54)
cat(sprintf("mean: %f\n", mean(probabilities)))
cat(sprintf("correlation: %f\n", cor(indices, probabilities)))
plot(indices, probabilities)
```
